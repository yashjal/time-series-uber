```python
import pandas as pd

df = pd.read_csv('data.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

f = df[["Pickup_date"]].copy()
f["Pickup_date"] = pd.to_datetime(f["Pickup_date"])
f.groupby(f["Pickup_date"].dt.month).count().plot(kind="bar")
plt.show()
```

```python
f.groupby(f["Pickup_date"].dt.weekday).count().plot(kind="bar")
plt.show()
```

```python
f.groupby(f["Pickup_date"].dt.hour).count().plot(kind="bar")
plt.show()
```

```python
df2 = pd.read_csv('data2.csv')
df2.head()
```

```python
f['Pickup_date_2014'] = df2['Date/Time']
f.head()
```

```python
f['Pickup_date_2014'] = pd.to_datetime(f['Pickup_date_2014'], format="%m/%d/%Y %H:%M:%S", errors='coerce')
f.tail()
```

```python
f.groupby(f["Pickup_date"].dt.weekday).count().plot(kind="bar")
f.groupby(f["Pickup_date"].dt.hour).count().plot(kind="bar")
plt.show()
```

```python
g = df2[["Date/Time"]].copy()
g['Date/Time'] = pd.to_datetime(g['Date/Time'], format="%m/%d/%Y %H:%M:%S", errors='coerce')
g.groupby(g["Date/Time"].dt.month).count().plot(kind="bar")
plt.show()
```
