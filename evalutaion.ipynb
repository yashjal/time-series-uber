```python
import pandas as pd

forecast = pd.read_csv('preds-2.csv')
forecast = forecast.drop(['Unnamed: 0'],axis=1)
forecast.head()
```

```python
forecast
```

```python
tmp = list(forecast.columns.values)
areas = []
for area in tmp:
    if not(area.endswith('lower') or area.endswith('upper')):
        areas.append(area)

areas.remove('ds')
areas.remove('trend')
areas.remove('Holiday')
areas.remove('Humidity')
areas.remove('Pressure')
areas.remove('Temperature')
areas.remove('Wind Direction')
areas.remove('Wind Speed')
areas.remove('daily')
areas.remove('extra_regressors')
areas.remove('rain')
areas.remove('seasonal')
areas.remove('seasonalities')
areas.remove('snow')
areas.remove('weekly')
areas.remove('yhat')

areas
```

```python
original = pd.read_csv('final-2.csv').tail(240522).reset_index(drop=True)
original = original.drop(['Unnamed: 0'],axis=1)

original
```

```python
p = forecast.loc[forecast['6_Alphabet City'] != 0.0]
o = original.loc[original['6_Alphabet City'] != 0]
p
```

```python
o
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
import numpy as np

count = 0
sum_error = 0
sum_abs_error = 0
#evaluation with yhat
for area in areas:
    p = forecast.loc[forecast[area] != 0.0]
    o = original.loc[original[area] != 0]
    comp = o[['y']]
    comp['yhat'] = p['yhat'].values
    comp[comp < 0] = 0
    comp_ar = comp.values
    if len(comp_ar) == 0:
        continue
    y_true = comp_ar[:,0]
    y_pred = comp_ar[:,1]
    sum_error += mean_squared_error(y_true, y_pred)
    sum_abs_error += mean_absolute_error(y_true, y_pred)
    count += 1

mean_error = sum_error/count
mean_abs_error = sum_abs_error/count
```

```python
#get entries with specific date TEST
original['ds'] = pd.to_datetime(original['ds'])  
mask = (original['ds'] > '2015-06-29') & (original['ds'] < '2015-07-01')
spec_date = original.loc[mask]
spec_date
```

```python
#getting specific time
index = pd.DatetimeIndex(spec_date['ds'])
spec_time = spec_date.iloc[index.indexer_between_time('17:00','17:30')]
```

```python
tmp = spec_time.loc[spec_time['6_Alphabet City'] != 0]
tmp.iloc[0]['y']
```

```python
x = []
y = []
y_upper = []
y_lower = []

for area in areas:
    p = spec_time.loc[spec_time[area] != 0.0]
    if p.empty:
        continue
    y.append(p.iloc[0]['y'])
    y_upper.append(p.iloc[0]['y'])
    y_lower.append(p.iloc[0]['y'])
    x.append(area)
```

```python
import matplotlib.pyplot as plt

nei = np.array(x)
pickups = np.array(y)
inds = pickups.argsort()
sorted_nei = nei[inds]
sorted_pickups = pickups[inds]
sorted_nei = sorted_nei[-20:]
sorted_pickups = sorted_pickups[-20:]


plt.xticks(rotation='vertical')
plt.plot(sorted_nei,sorted_pickups,'o')
plt.show()
```
