```python
import pandas as pd
#read
df = pd.read_csv('final-2.csv')
df=df.drop(['Unnamed: 0'],axis=1)
df.tail(210000)
```

```python
df.tail(240522)
```

```python
from fbprophet import Prophet
#initiate
m = Prophet()
```

```python
#add regressors
cols=list(df.columns.values)[2:]
for col in cols:
    m.add_regressor(col)
```

```python
#divide into train,valid
train=df.loc[0:(1322691-7651-1),:]
train
```

```python
#valid=all 7651 entries from last day, i.e. 30th June 2015
cols=list(df.columns.values)
cols.remove('y')
valid=df.loc[1315040:,cols]
valid
```

```python
#train/fit model
m.fit(train)
```

```python
#predict
forecast = m.predict(valid)
forecast
```

```python
#save predictions
forecast.to_csv('preds.csv')
```
