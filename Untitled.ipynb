```python
import pandas as pd
import numpy as np
from fbprophet import Prophet

df = pd.read_csv('mod_manhattan_set.csv')
df.head()
```

```python
df.drop(['Unnamed: 0','0','1','3','4','5'])
df
```

```python
g = df.groupby('6', as_index=False)
d = dict(iter(g))
g.get_group('Alphabet City')
```

```python
df_new = pd.DataFrame({'2':[],'6':[],'7':[],'8':[],'9':[]})

uni = df['6'].unique()
for neighborhood in uni:
    tmp = g.get_group(neighborhood)
    nei = tmp.groupby('8', as_index=False)
    uni2 = tmp['8'].unique()
    for bucket in uni2:
        tmp2 = nei.get_group(bucket)
        numPicks = len(tmp2)
        rep = tmp2.iloc[[numPicks/2]]
        rep['9'] = numPicks
        df_new.append(rep)

df_new
```
