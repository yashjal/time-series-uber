```python
import pandas as pd

df1 = pd.read_csv('boro/statenIsland_final.csv')
df2 = pd.read_csv('boro/manhattan_final.csv')
df3 = pd.read_csv('boro/brooklyn_final.csv')
df4 = pd.read_csv('boro/bronx_final.csv')
df5 = pd.read_csv('boro/queens_final.csv')

df = pd.concat([df1,df2,df3,df4,df5])
df
```

```python
df.loc[df['7_Arden Heights'] == 1]
```

```python
df.fillna(0, inplace=True)
df
```

```python
df=df.drop(['Unnamed: 0'],axis=1)
df
```

```python
cols = df.columns.tolist()
tmp = cols[-1]
tmp2 = cols[-2]
cols[-1] = cols[1]
cols[-2] = cols[0]
cols[0] = tmp2
cols[1] = tmp

df=df[cols]
df
```

```python
df['Date'] = pd.to_datetime(df['ds'])
df=df.sort_values(by='Date')
df=df.drop(['Date'],axis=1)
df
```

```python
df.to_csv('locations_final_no_weather_hols.csv')
```
