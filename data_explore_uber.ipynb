```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('data.csv')
df.head()
```

```python
df2 = pd.read_csv('data2.csv')
df2.head()
```

```python
f = df[["Pickup_date"]].copy()
f["Pickup_date"] = pd.to_datetime(f["Pickup_date"])

f2 = df2[["Date/Time"]].copy()
f2["Date/Time"] = pd.to_datetime(f2["Date/Time"], format="%m/%d/%Y %H:%M:%S", errors='coerce')
```

```python
f
```

```python
#get entries by week day
f['weekday'] = f['Pickup_date'].dt.dayofweek
f.loc[f['weekday'] == 0]
```

```python
mon = f.loc[f['weekday'] == 0]
mon.drop(['weekday'],axis=1,inplace=True)
tues = f.loc[f['weekday'] == 1]
tues.drop(['weekday'],axis=1,inplace=True)
wed = f.loc[f['weekday'] == 2]
wed.drop(['weekday'],axis=1,inplace=True)
thurs = f.loc[f['weekday'] == 3]
thurs.drop(['weekday'],axis=1,inplace=True)
fri = f.loc[f['weekday'] == 4]
fri.drop(['weekday'],axis=1,inplace=True)
sat = f.loc[f['weekday'] == 5]
sat.drop(['weekday'],axis=1,inplace=True)
sun = f.loc[f['weekday'] == 6]
sun.drop(['weekday'],axis=1,inplace=True)

mon
```

```python
x0=mon.groupby(mon["Pickup_date"].dt.hour).count()
x0['tuesday']=tues.groupby(tues["Pickup_date"].dt.hour).count()
x0['wednesday']=wed.groupby(wed["Pickup_date"].dt.hour).count()
x0['thursday']=thurs.groupby(thurs["Pickup_date"].dt.hour).count()
x0['friday']=fri.groupby(fri["Pickup_date"].dt.hour).count()
x0['saturday']=sat.groupby(sat["Pickup_date"].dt.hour).count()
x0['sunday']=sun.groupby(sun["Pickup_date"].dt.hour).count()
x0.rename(index=str, columns={"Pickup_date": "monday"},inplace=True)

x0
```

```python
x0.plot()
plt.xlabel("hour of the day")
plt.ylabel("# pickups")
plt.show()
```

```python

tues.groupby(tues["Pickup_date"].dt.hour).count().plot()
wed.groupby(wed["Pickup_date"].dt.hour).count().plot()
thurs.groupby(thurs["Pickup_date"].dt.hour).count().plot()
fri.groupby(fri["Pickup_date"].dt.hour).count().plot()
sat.groupby(sat["Pickup_date"].dt.hour).count().plot()
sun.groupby(sun["Pickup_date"].dt.hour).count().plot()

plt.xlabel("hour of the day")
plt.ylabel("# pickups")
plt.legend(["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"])
plt.show()
```

```python
f2.groupby(f2["Date/Time"].dt.month).count().plot(kind="bar")
plt.xlabel("month")
plt.ylabel("# pickups")
plt.legend(["2014 data"])

f.groupby(f["Pickup_date"].dt.month).count().plot(kind="bar",color='r')
plt.xlabel("month")
plt.ylabel("# pickups")
plt.legend(["2015 data"])

plt.show()
```
