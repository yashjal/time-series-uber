```python
import pandas as pd
#read
df = pd.read_csv('final.csv')
df=df.drop(['Unnamed: 0'],axis=1)

list(df.columns.values)
```

```python
df['Weather Type'].unique()
```

```python
df['Weather Type 2'] = df['Weather Type']
list(df.columns.values)
```

```python
def snow(x):
    if x == 'light snow':
        return 1
    elif x == 'snow':
        return 2
    else:
        return 0

def rain(x):
    if x == 'light intensity drizzle' or x == 'drizzle':
        return 1
    elif x == 'light rain' or x == 'proximity thunderstorm':
        return 2
    elif x == 'moderate rain' or x == 'thunderstorm with light rain':
        return 3
    elif x == 'heavy intensity rain' or x == 'very heavy rain' or x == 'thunderstorm':
        return 4
    else:
        return 0
```

```python
df['Weather Type']=df['Weather Type'].apply(snow)
df['Weather Type 2']=df['Weather Type 2'].apply(rain)

df['Weather Type'].unique()
```

```python
df['Weather Type 2'].unique()
```

```python
df=df.rename(index=str, columns={"Weather Type": "snow", "Weather Type 2": "rain"})
list(df.columns.values)
```

```python
cols = df.columns.tolist()
tmp = cols[-1]
cols[-1] = cols[-2]
cols[-2] = tmp

df=df[cols]
list(df.columns.values)
```

```python
df.to_csv('final-2.csv')
```
